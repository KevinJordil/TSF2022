```python
import numpy as np
import matplotlib.pyplot as plt
```

### Wifi power

```python
d = np.arange(1, 32, 0.1)

delta = 3e8 / 5.8e9
# Formule de Friis
#    Pt  Gt  Gr  
Pr = 14 + 1 + 1 + 20 * np.log10(delta) - 20 * np.log10(4 * np.pi * d)

# obstacle
# 5.8 GHz
# Interieur avec obstacle
# 14dBm
Pr_ob = 14 + 1 + 1 + 20 * np.log10(delta) - 20 * np.log10(4 * np.pi) - 50 * np.log10(d)

plt.plot(d, Pr_ob, label="Avec obstacle")
plt.plot(d, Pr, label="Sans obstacle")
plt.axhline(y = -90, color = 'b', linestyle = ':')
plt.grid()
plt.xlabel("Distance [m]")
plt.ylabel("Puissance re√ßue [dMb]")
plt.legend()
```
